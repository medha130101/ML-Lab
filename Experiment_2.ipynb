```python
import numpy as np
corr = df.corr().round(2)
mask = np.zeros_like(corr,dtype = np.bool)
mask[np.triu_indices_from(mask)] = True
# Set the size of figure
f,ax = plt.subplots(figsize = (20,20))
# Define Custom Colormap
cmap = sns.diverging_palette(220,10,as_cmap = True)
# Draw the HeatMap
sns.heatmap(corr,mask = mask,cmap = "crest",vmin = -1, vmax = 1, center = 0,square = True,linewidths = .5, cbar_kws = {"shrink":0.5},annot=True)
plt.tight_layout()
```

```python
# Normalization
# Drop the 'diagnosis' column as it consists of string outputs.
x = df.drop('diagnosis',axis=1)
x
```

```python
x = (x - np.min(x))/ (np.max(x) - np.min(x))
x
```

```python
y  = df[['diagnosis']]
print(y)
```

```python
# Split the dataset into train and test data sets.
from sklearn.model_selection import train_test_split
# For checking the results.
from sklearn.metrics import classification_report, confusion_matrix
# To visulaise the output.
from sklearn.tree import plot_tree
# The train-test split is a 80-20 split for the case.
x_train , x_test , y_train , y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
print( " Training Input Shape",x_train.shape)
print( " Test Input Shape" , x_test.shape)


```

```python
# Applying the Naive bayes Algorithm
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(x_train, y_train.values.ravel())
print("Naive Bayes Score:", nb.score(x_test,y_test))
```

```python
y_pred  = nb.predict(x_test)
```

```python
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
plt.figure(figsize = (5,5))
sns.heatmap(data = cm , linewidths = 1.0 , square = True , annot = True, cmap = 'Wistia')
plt.ylabel ( " Actual Label")
plt.xlabel ( " Predicted Label")
```
